```python
import pybamm
import ipywidgets as widgets
```

```python
# Create a list of models to compare.
models = [
    pybamm.lithium_ion.DFN(),
    pybamm.lithium_ion.SPMe(),
    pybamm.lithium_ion.SPM()
]
```

```python
# Loop the model list, solve simulations and store in "sims".
sims =[]
for model in models:
    sim = pybamm.Simulation(model)
    sim.solve([0, 3600])  # Solve for 1 hour
    sims.append(sim)
```

```python
# Pass the list of simulations to the dynamic plot function, which can plot different solutions in the same figures. 
pybamm.dynamic_plot(sims)
```
