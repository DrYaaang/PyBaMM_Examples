```python
import pybamm
import ipywidgets as widgets
```

```python
# Choose Doyle-Fuller-Newman (DFN) model (P2D model)
model = pybamm.lithium_ion.DFN()
```

```python
# Use the model to create a PyBaMM simulation
sim = pybamm.Simulation(model)
```

```python
# Solve for 1 h
sim.solve([0, 3600])
```

```python
#  Generate a dynamic plot of the key variables
sim.plot()
```
