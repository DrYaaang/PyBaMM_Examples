```python
import pybamm
```

```python
# options:
# lumped: One cell temperature changes with time, but is uniform in space.
# isothermal: No temperature change. Temperature is fixed.
# X-full: Temperature changes with both time and position through the cell thickness.
model = pybamm.lithium_ion.DFN(options={"thermal": "x-full"})  # Choose options

sim = pybamm.Simulation(model)
sim.solve([0, 3600])
```

```python
parameter_values.update({
    "Current function [A]": 20,
    "Negative electrode thermal conductivity [W.m-1.K-1]": 0.0005,
    "Separator thermal conductivity [W.m-1.K-1]": 0.01,
    "Positive electrode thermal conductivity [W.m-1.K-1]": 0.0005,
})

sim = pybamm.Simulation(model, parameter_values=parameter_values)
sim.solve([0, 3600])
```

```python
sim.plot(
    ["Cell temperature [K]", "Total heating [W.m-3]", "Current [A]", "Voltage [V]"]
)
```

```python
model = pybamm.lithium_ion.SPMe(options={"thermal": "x-full"})

parameter_values = model.default_parameter_values.copy()

parameter_values.update({
    "Current function [A]": 20,
    "Negative electrode thermal conductivity [W.m-1.K-1]": 0.005,
    "Separator thermal conductivity [W.m-1.K-1]": 0.01,
    "Positive electrode thermal conductivity [W.m-1.K-1]": 0.005,
})

sim = pybamm.Simulation(model, parameter_values=parameter_values)
sol = sim.solve([0, 300])

sol.plot(["Cell temperature [K]", "Total heating [W.m-3]"])
```
